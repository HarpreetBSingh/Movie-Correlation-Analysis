```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('movies.csv')
```

```python
# Head of df
df.head()
```

```python
# Viewing missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Checking data types

df.dtypes
```

```python
# Changing data type of budget and gross, from floats into ints

df['budget'] = df['budget'].astype('int64', errors = 'ignore')

df['gross'] = df['gross'].astype('int64', errors = 'ignore')
```

```python
# Released and year are different in some cases, one could have a year of 1986 but release of 1987
# For sake of easier calculations, we want to make a new column that is the same as released
df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
# Sort by gross descending
df.sort_values(by=['gross'],inplace = False, ascending=False)
```

```python
# Check for duplicates and drop
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot to compare budget to gross revenue
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Film Budget')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
df.head()
```

```python
# Plotting budget vs gross revenue using seaborn's regression plot

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color": "red"}, line_kws={"color":"green"})
```

```python
# We see a positive correlation between budget and gross revenue, but it is difficult saying how much
df.corr()
```

Budget and gross has a correlation of .740395 which is above .7 and displays strong correlation


```python
# Visualizing correlation with a correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Quantitative Features')
plt.xlabel('Film Features')
plt.ylabel('Film Features')

plt.show()
```

Anything that is black has a lower correlation, however, the brighter colours indicate a higher correlation

```python
# Convert company to a numeric representation, in order to run correlation analysis

df_convert = df

for col_name in df_convert.columns:
    if(df_convert[col_name].dtype == 'object'):
        df_convert[col_name] = df_convert[col_name].astype('category')
        df_convert[col_name] = df_convert[col_name].cat.codes
        
df_convert
```

```python
# Visualizing correlation with a correlation matrix for all features

correlation_matrix = df_convert.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Quantitative Features')
plt.xlabel('Film Features')
plt.ylabel('Film Features')

plt.show()
```

```python
df_convert.corr()
```

```python
# Using unstack command to see correlations 

corr_mat = df_convert.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > .5]

high_corr
```

Both votes and budget have a high correlation with gross revenues
